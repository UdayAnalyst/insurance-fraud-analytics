#Assumption: #
Incident-level and reporting variables may be missing at claim submission time. Missing values were retained and handled via imputation within the modeling pipeline, allowing the model to learn from missingness patterns without introducing data leakage.

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns
categorical_features = df.select_dtypes(include=['object']).columns
```

```python
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean'))
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessing = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```

```python
drop_cols = [
    'policy_number',
    'insured_zip',
    'incident_location',
    'auto_model',
    'policy_bind_date',
    'incident_date',
    'injury_claim',
    'property_claim',
    'vehicle_claim',
    'capital-gains',
    'capital-loss'
]

df_model = df.drop(columns=drop_cols)
```

I have explicitly removed identifier and high-cardinality fields such as policy numbers, ZIP codes, and exact locations to prevent memorization and bias.
```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression

numeric_features = X.select_dtypes(include=['int64','float64']).columns
categorical_features = X.select_dtypes(include=['object']).columns

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)

model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(
        max_iter=1000,
        class_weight='balanced',
        solver='lbfgs'
    ))
])
```

```python
import numpy as np
import pandas as pd

target = "fraud_reported"

# Ensure df_model exists and is what you're using
df_model = df_model.copy()

# Convert all pandas nullable NA to numpy NaN
df_model = df_model.replace({pd.NA: np.nan})

# (Optional but recommended) also convert literal '?' if any remain
df_model = df_model.replace("?", np.nan)

# Build X/y fresh from this cleaned df_model
X = df_model.drop(columns=[target])
y = df_model[target].astype(int)

# Train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=42
)

# Sanity check: confirm no pd.NA objects remain
print("Any pd.NA in X_train?", X_train.applymap(lambda v: v is pd.NA).any().any())
print("Missing values (top 5 cols):")
print(X_train.isna().sum().sort_values(ascending=False).head(5))
```

```python
import numpy as np
import pandas as pd

target = "fraud_reported"

X = df_model.drop(columns=[target]).copy()
y = df_model[target].astype(int)
```

```python
cat_cols = X.select_dtypes(include=["object"]).columns.tolist()

X[cat_cols] = X[cat_cols].fillna("__MISSING__").astype(str)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=42
)
```

```python
cat_feature_indices = [X.columns.get_loc(c) for c in cat_cols]
```

```python
from catboost import CatBoostClassifier

# weight fraud class higher
pos_weight = (len(y_train) - y_train.sum()) / y_train.sum()

cat_model = CatBoostClassifier(
    iterations=800,
    learning_rate=0.05,
    depth=6,
    loss_function="Logloss",
    eval_metric="AUC",
    random_seed=42,
    verbose=100,
    class_weights=[1, pos_weight]
)

cat_model.fit(X_train, y_train, cat_features=cat_feature_indices)
```

CatBoost requires categorical values to be strings or integers, so I encoded missing categorical values as a dedicated â€˜MISSINGâ€™ category while leaving numeric NaNs untouched.

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, average_precision_score

cat_pred = cat_model.predict(X_test)
cat_prob = cat_model.predict_proba(X_test)[:, 1]

print("Confusion matrix:\n", confusion_matrix(y_test, cat_pred))
print("\nReport:\n", classification_report(y_test, cat_pred, digits=3))
print("\nROC-AUC:", roc_auc_score(y_test, cat_prob))
print("PR-AUC:", average_precision_score(y_test, cat_prob))
```

I compared an interpretable logistic regression, a Random Forest, and CatBoost. While Random Forest improved ranking, CatBoost delivered the best overall balanceâ€”achieving the highest PR-AUC while maintaining strong fraud recall and improved precision. This makes CatBoost the most suitable model for operational fraud triage

Investigators have limited capacity, so instead of using a fixed 0.5 probability threshold, we prioritize the top-K highest-risk claims.

```python
cat_prob = cat_model.predict_proba(X_test)[:, 1]
```

If we only review the top K% riskiest claims, how much fraud do we catch?

```python
import numpy as np
from sklearn.metrics import confusion_matrix

def evaluate_top_k(y_true, y_prob, k_percent):
    threshold = np.percentile(y_prob, 100 - k_percent)
    y_pred_k = (y_prob >= threshold).astype(int)

    tn, fp, fn, tp = confusion_matrix(y_true, y_pred_k).ravel()

    recall = tp / (tp + fn)
    precision = tp / (tp + fp)
    review_rate = (tp + fp) / len(y_true)

    return {
        "Top_K_%": k_percent,
        "Threshold": threshold,
        "Recall": recall,
        "Precision": precision,
        "Claims_Reviewed_%": review_rate * 100,
        "Fraud_Caught": tp,
        "Fraud_Missed": fn
    }
```

These are realistic investigator capacities.

```python
results = []
for k in [10, 15, 20, 25, 30]:
    results.append(evaluate_top_k(y_test, cat_prob, k))

import pandas as pd
pd.DataFrame(results)
```

Based on a Top-K analysis, a 25% review threshold provides the best balance between fraud detection and operational capacity, capturing approximately 65% of fraudulent claims while maintaining precision above 63%

By reviewing only one quarter of claims, the model enables investigators to identify nearly two-thirds of fraud while minimizing unnecessary reviews.

Ensemble (Soft Voting)

```python
lr_prob = model.predict_proba(X_test)[:, 1]
rf_prob = rf_model.predict_proba(X_test)[:, 1]
cb_prob = cat_model.predict_proba(X_test)[:, 1]
```

```python
ens_prob = 0.3*lr_prob + 0.2*rf_prob + 0.5*cb_prob
```

```python
from sklearn.metrics import roc_auc_score, average_precision_score, confusion_matrix, classification_report

ens_pred = (ens_prob >= 0.5).astype(int)

print("Confusion matrix:\n", confusion_matrix(y_test, ens_pred))
print("\nReport:\n", classification_report(y_test, ens_pred, digits=3))

print("\nROC-AUC:", roc_auc_score(y_test, ens_prob))
print("PR-AUC:", average_precision_score(y_test, ens_prob))
```

```python
results_ens = []
for k in [10, 15, 20, 25, 30]:
    results_ens.append(evaluate_top_k(y_test, ens_prob, k))

import pandas as pd
pd.DataFrame(results_ens)
```

I used a probability-based ensemble to combine complementary model strengthsâ€”CatBoost for categorical interactions, logistic regression for stability/calibration, and Random Forest for nonlinear patterns. I evaluated the ensemble using PR-AUC and Top-K recall to match investigator capacity

experimented with a soft-voting ensemble combining logistic regression, Random Forest, and CatBoost. While the ensemble slightly improved ROC-AUC, it did not improve PR-AUC or Top-K fraud capture compared to CatBoost alone. Given the added complexity and lack of operational lift, CatBoost was selected as the final model
